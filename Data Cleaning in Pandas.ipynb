```python
import pandas as pd
```

```python
df = pd.read_excel(r'D:\FILE DATA ANALYST\Belajar Jupyter Notebook\Data Cleaning Using Pandas\Customer Call List.xlsx')
df
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop(columns = 'Not_Useful_Column')
```

```python
df
```

```python
# df['Last_Name'] = df['Last_Name'].str.strip('...')
# df['Last_Name'] = df['Last_Name'].str.strip('_')
# df['Last_Name'] = df['Last_Name'].str.strip('/')
df['Last_Name'] = df['Last_Name'].str.strip('123._/')
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-z0-9]','')
#cleaning phone number with Regex 
#taking everything that is not abc, 123 (letter or number)
df
```

```python
#If we want to change the type of phone number to str
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : str(x))
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na','')
```

```python
df
```

```python
df[['Street Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', 2, expand=True) #the split will give us 2 new column to save the splitted string
#we need to put it in the new column
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.lower().replace('yes', 'y')
df['Paying Customer'] = df['Paying Customer'].str.lower().replace('no', 'n')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.lower().replace('yes', 'y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.lower().replace('no', 'n')
```

```python
df
```

```python
df = df.fillna('')
df = df.replace('N/a', '')
df = df.replace('n/a', '')
df = df.fillna('')
df
```

# We can also drop/remove row that doesn't has phone number or do not want to be contacted

```python
can_be_contacted_df = df.copy(deep=True)
for x in can_be_contacted_df.index:
    if can_be_contacted_df.loc[x, 'Do_Not_Contact'] == 'y' or can_be_contacted_df.loc[x, 'Phone_Number'] == '':
        can_be_contacted_df.drop(x, inplace=True)
        
can_be_contacted_df
#another way to remove null values: can_be_contacted_df = can_be_contacted_df.dropna(subset='Phone_Number', inplace=True)
```

```python
can_be_contacted_df = can_be_contacted_df.reset_index(drop=True)
```
